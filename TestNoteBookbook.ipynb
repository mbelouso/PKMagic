# Test code for PK Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
pkdata = pd.read_csv('warfpk.csv')

# Display the first few rows of the DataFrame
print(pkdata.head())
```

```python
# Ensure expected input exists
if "pkdata" not in globals():
    raise NameError("pkdata is not defined. Run the data-loading cell first.")
if "#ID" not in pkdata.columns:
    raise KeyError("Column '#ID' was not found in pkdata.")

# Split into one DataFrame per patient ID
patient_dfs = {
    patient_id: group.reset_index(drop=True)
    for patient_id, group in pkdata.groupby("#ID")
}

print(f"Created {len(patient_dfs)} patient-specific DataFrames.")
print("Sample patient IDs:", list(patient_dfs.keys())[:5])

# Example access for analysis
first_id = next(iter(patient_dfs))
print(f"\nPreview for patient {first_id}:")
print(patient_dfs[first_id].head())
```
